## This notebook converts raw which is present in MS excel sheets to csv files.

```python
import pandas as pd
from os import path
```

### Loading yearly $CO_{2}$ emmision data from excel files into dataframe

```python
co2_year_data_file_path = path.join("raw_data", "carbon_dioxide_data","CO2_YEARLY_DATA_1970-2021.xlsx")
df_co2_yearly_ipcc_2006 = pd.read_excel(co2_year_data_file_path, sheet_name= 0, usecols="A:BH", skiprows=10)
df_co2_yearly_ipcc_1996 = pd.read_excel(co2_year_data_file_path, sheet_name= 1, usecols="A:BH", skiprows=10)
df_co2_yearly_total = pd.read_excel(co2_year_data_file_path, sheet_name= 2, usecols="A:BE", skiprows=10)
```

### Saving yearly $CO_{2}$ emmision data in csv files

```python
csv_dir_path = path.join("data","carbon_dioxide")
df_co2_yearly_ipcc_1996.to_csv(path.join(csv_dir_path, "co2_yearly_ipcc_1996.csv"), index= False)
df_co2_yearly_ipcc_2006.to_csv(path.join(csv_dir_path, "co2_yearly_ipcc_2006.csv"),index= False)
df_co2_yearly_total.to_csv(path.join(csv_dir_path, "co2_yearly_ipcc_total.csv"),index= False)
```

### Loading monthly $CO_{2}$ emmision data from excel file in dataframe

```python
co2_monthly_data_file_path = path.join("raw_data", "carbon_dioxide_data","CO2_Monthly_2000_2021.xlsx")
df_co2_monthly_ipcc_2006 = pd.read_excel(co2_monthly_data_file_path, sheet_name= 0, usecols="A:U", skiprows=9)
df_co2_monthly_ipcc_1996 = pd.read_excel(co2_monthly_data_file_path, sheet_name= 1, usecols="A:U", skiprows=9)
df_co2_monthly_total = pd.read_excel(co2_monthly_data_file_path, sheet_name= 2, usecols="A:R", skiprows=9)
```

### Saving monthly $CO_{2}$ emmision data in csv files

```python
csv_dir_path = path.join("data","carbon_dioxide")
df_co2_monthly_ipcc_1996.to_csv(path.join(csv_dir_path, "co2_monthly_ipcc_1996.csv"), index= False)
df_co2_monthly_ipcc_2006.to_csv(path.join(csv_dir_path, "co2_monthly_ipcc_2006.csv"),index= False)
df_co2_monthly_total.to_csv(path.join(csv_dir_path, "co2_monthly_ipcc_total.csv"),index= False)
```
